### I.2 Chaîne de Markov

```python
#Création du dictionnaire clé-valeur associant états et caractères

dic={'1' : ' ', '2' : 'a', '3' : 'b', '4': 'c', '5' : 'd', '6' : 'e', '7': 'f', '8' : 'g', '9' : 'h', '10': 'i', '11': 'j',
'12' : 'k', '13' : 'l', '14': 'm', '15' : 'n', '16' : 'o', '17': 'p', '18' : 'q', '19' : 'r' , '20': 's', '21' : 't', '22'
: 'u', '23': 'v', '24' : 'w', '25' : 'x' , '26': 'y', '27' : 'z', '28' : ' ' }
```

<blockquote> La première ligne de la matrice de transition correspond à la probabilité d'aller dans n'importe quel état, étant donné que vous êtes actuellement dans le premier état.<br><br>

Celles de la dernière colonne correspondent à la probabilité d'aller au dernier état, pour tous les états possibles dans lesquels vous pouvez être actuellement.

```python
#ouvrez le fichier et lisez-le ligne par ligne
file = open('bigramenglish.txt').read().splitlines()
trans_matrix = np.zeros((28,28))

# construire la matrice
for i in range (len(file)):
    item = file[i].split()
    for j in range (len(item)):
        trans_matrix[i][j] = item[j]

letter = 1
for prob in trans_matrix:
    indice = np.argmax(prob)
    print("- La transition la plus fréquente depuis de ",dic.get(str(letter)), "(",letter,"), est:", dic.get(str(indice+1)),"(",indice+1,").\n")
    letter +=1
```

```python
def etat_suivant(state,trans_matrix):
    prob = np.random.random()
    l = trans_matrix[state-1]
    cum_sum = np.cumsum(l)
    next_state = 1
    for i in cum_sum:
        if prob > i:
            next_state += 1
        else:
            break
    return next_state
print(etat_suivant(1,trans_matrix))
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot()
fig.subplots_adjust(top=0.85)

#plt.figure(figsize=(10, 6))
ax.plot(np.arange(1,29,1), np.cumsum(trans_matrix[1]))
ax.scatter(np.arange(1,29,1), np.cumsum(trans_matrix[1]),marker="*")
ax.axhline(0.69,color="k")
ax.axvline(19.6,color="k")
ax.set_xlabel("états")
ax.set_ylabel("fonction de répartition (cumsum)")
ax.set_title("Générer l'état suivant à partir de l'état 1")
ax.text(0.77, 0.01, 'Q(t+1)',
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax.transAxes, fontsize=15)
ax.text(0.01, 0.68, 'Rand',
        verticalalignment='bottom', horizontalalignment='left',
        transform=ax.transAxes, fontsize=15)

plt.show()
```

<blockquote> Pour générer l'état suivant, nous tirons une valeur aléatoire entre 0 et 1. Ensuite, l'état suivant correspondra à l'état dans la plage de la cumsum probabilité, comme affiché dans le graphique ci-dessus.

```python
def genere_state_seq(trans_matrix):
    state = 1
    seq = [1]
    while state != trans_matrix.shape[0]:
        next_state = etat_suivant(state,trans_matrix)
        seq.append(next_state)
        state = next_state      
    return seq


def display_seq(seq,dic):
    mot=[]
    for i in seq:
        mot.append(dic.get(str(i)))
    return ''.join(mot)
    
```

```python
# Exemple de génération de 10 mots
for i in range(10):
    seq = genere_state_seq(trans_matrix)
    print(display_seq(seq,dic))
```

```python
def modifie_mat_dic (old_dict, old_matrix, new_state, prob_from_state, prob_to_state):
    if sum(prob_from_state)!=1:
        print("La somme des probabilités doit être 1")
     
    num_states= len(old_dict)
    new_dict = copy.deepcopy(old_dict)
    new_dict[str(num_states+1)] = new_state
    
    new_matrix = np.copy(old_matrix)
    new_matrix = np.column_stack((new_matrix, prob_to_state[0:num_states]))
    new_matrix = np.vstack((new_matrix, prob_from_state))
    new_matrix[num_states-1,num_states-1] = 0
    new_matrix[num_states-1,0] = old_matrix[num_states-1,num_states-1] - new_matrix[num_states-1,num_states]
    
    return new_matrix, new_dict
```

```python
prob_from_state = np.zeros(29)
prob_from_state[28] = 1
prob_to_state = np.zeros(29)
prob_to_state[27] = 0.1
trans_matrix2, dic2 = modifie_mat_dic (dic, trans_matrix, ".", prob_from_state, prob_to_state)

print(dic2)
print(trans_matrix2)
```

```python
# Exemple de génération d'une phrase
for i in range(10):
    seq = genere_state_seq(trans_matrix2)
    ph = display_seq(seq,dic2)
    print(ph,"\n")
```

```python
#ouvrez le fichier et lisez-le ligne par ligne
file = open('bigramfrancais.txt').read().splitlines()
trans_matrix_fr = np.zeros((28,28))

# construire la matrice
for i in range (len(file)):
    item = file[i].split()
    for j in range (len(item)):
        trans_matrix[i][j] = item[j]

prob_from_state = np.zeros(29)
prob_from_state[28] = 1
prob_to_state = np.zeros(29)
prob_to_state[27] = 0.1
trans_matrix_fr2, dic2 = modifie_mat_dic (dic, trans_matrix, ".", prob_from_state, prob_to_state)
```

```python
def calc_vraisemblance(phrase,trans_matrix,not_print=True):    
    words = phrase.split()
    vraisemblance = 1.0
    
    # list out keys and values separately
    key_list = list(dic2.keys())
    val_list = list(dic2.values())

    for w in words:
        if (w!=words[-1]):
            p = val_list.index(w[0])
            j = int(key_list[p])-1
            vraisemblance = vraisemblance*trans_matrix[0][j]
            if (not not_print): print("-/",w[0],":",trans_matrix[0][j])
            for char in range(len(w)-1):
                p1 = val_list.index(w[char])
                p2 = val_list.index(w[char+1])
                i = int(key_list[p1])-1
                j = int(key_list[p2])-1
                if (not not_print):  print(w[char],"/", w[char+1], ":", trans_matrix[i][j])
                vraisemblance = vraisemblance*trans_matrix[i][j]

            vraisemblance = vraisemblance * trans_matrix[j][27]
            vraisemblance = vraisemblance  * 0.9
            if (not not_print):  print(w[-1],"/+:",trans_matrix[j][27],"\n")
            if (w!=words[-2]):
                if (not not_print):  print("+/-:",0.9,"\n")
        else:
            vraisemblance = vraisemblance * 0.1
            if (not not_print):  print("+/.:",0.1)
            
    return vraisemblance
```

```python
vr_ing_ing =calc_vraisemblance("to be or not to be .",trans_matrix2,False)
vr_ing_fr =calc_vraisemblance("to be or not to be .",trans_matrix_fr2)
vr_fr_ing =calc_vraisemblance("etre ou ne pas etre .",trans_matrix2)
vr_fr_fr =calc_vraisemblance("etre ou ne pas etre .",trans_matrix_fr2)
```

```python
print("\n\nVraisemblance modèles anglais (to be or not to be.)", vr_ing_ing)
print("\n\nVraisemblance modèles français (to be or not to be.)", vr_ing_fr)
print("\n\nVraisemblance modèles anglais (etre ou nepas etre.)", vr_fr_ing)
print("\n\nVraisemblance modèles français (etre ou nepas etre.)", vr_fr_fr)
```

<blockquote> Comme prévu, la probabilité que la phrase anglaise soit générée suivant la probabilité de la langue anglaise est plus grande qu'une phrase anglaise par le français. Il en va de même pour l'inverse.
